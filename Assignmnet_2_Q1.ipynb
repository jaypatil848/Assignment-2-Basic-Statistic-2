```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
```

```python
title=pd.Series(['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert'])
```

```python
# Pie Plot
plt.figure(figsize=(10,6))
plt.pie(df,labels=title,autopct='%1.0f%%')
plt.show()
```

```python
# Box Plot to find outliars
plt.boxplot(df)
plt.title('Boxplot')
plt.show()
```

```python
# Mean
df.mean()
```

```python
# Vairance
df.var()
```
